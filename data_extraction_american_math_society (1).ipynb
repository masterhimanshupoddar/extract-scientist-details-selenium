```
import bs4 as bs
import json
import urllib.request
file = open("data.txt","a")


batch_sz = 100
count = 0   


id = 150001



while(id<=200000):
    #file_name = str(id)+".txt"
    #file = open(file_name,"w")
    count +=1
    if(count%200==0):
      print(count)
    if(count == batch_sz):
        count = 0
        file.close()
        file = open("data.txt","a")
    try:
        url = 'https://www.genealogy.math.ndsu.nodak.edu/id.php?id='+str(id)
        mkcntn = urllib.request.urlopen(url).read()
        soup = bs.BeautifulSoup(mkcntn,'html.parser')
        stud_name = soup.h2.text.strip("\n")
        div = soup.find('div',id = 'paddingWrapper')
        p = div.find_all('p')
        try:
            advisor = p[1].find_all('a')
            if(len(advisor)==2):
                advisor1_name = advisor[0].text
                advisor1_id = advisor[0].get('href')
                advisor1_id = advisor1_id.replace("id.php?id=","")
                advisor2_name = advisor[1].text
                advisor2_id = advisor[1].get('href')
                advisor2_id = advisor2_id.replace("id.php?id=","")
            else:
                advisor1_name = advisor[0].text
                advisor1_id = advisor[0].get('href')
                advisor1_id = advisor1_id.replace("id.php?id=","")
                advisor2_name = ""
                advisor2_id = ""
        except:
                advisor1_name = ""
                advisor1_id = ""
                advisor2_name = ""
                advisor2_id = ""
        school = div.find_all('span')
        try:
            university = school[1].text
        except:
            university  = ""
    except:
            print(id,"error")
            stud_name = ""
            university = ""
            advisor1_name = ""
            advisor1_id = ""
            advisor2_name = ""
            advisor2_id = ""
    main_dict = {
    'id' : str(id),
    'name' : stud_name,
    'university' : university,
    'advisor_1_name' : advisor1_name,
    'advisor_1_id' : advisor1_id,
    'advisor_2_name' : advisor2_name,
    'advisor_2_id' : advisor2_id
    }
    json_data = json.dumps(main_dict)
    file.write(json_data)
    file.write("\n")
    id=id+1
file.close()
```
